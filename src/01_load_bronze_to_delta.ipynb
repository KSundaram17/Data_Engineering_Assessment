```python
from config import load_config

cfg = load_config("config_databricks.yaml")
```

```python
df_jobs = spark.read.option("header", True).csv(cfg["bronze_jobs_export"])
df_jobs.write.format("delta").mode("overwrite").saveAsTable(cfg["bronze_jobs_table"])

df_candidates = spark.read.option("header", True).csv(cfg["bronze_candidates_export"])
df_candidates.write.format("delta").mode("overwrite").saveAsTable(cfg["bronze_candidates_table"])

df_education = spark.read.option("header", True).csv(cfg["bronze_education_export"])
df_education.write.format("delta").mode("overwrite").saveAsTable(cfg["bronze_education_table"])

df_applications = spark.read.option("header", True).csv(cfg["bronze_applications_export"])
df_applications.write.format("delta").mode("overwrite").saveAsTable(cfg["bronze_applications_table"])

df_workflow = spark.read.option("header", True).csv(cfg["bronze_workflow_events_export"])
df_workflow.write.format("delta").mode("overwrite").saveAsTable(cfg["bronze_workflow_events_table"])
```

```python
spark.sql("SELECT COUNT(*) FROM bronze_jobs").show()
spark.sql("SELECT COUNT(*) FROM bronze_candidates").show()
spark.sql("SELECT COUNT(*) FROM bronze_education").show()
spark.sql("SELECT COUNT(*) FROM bronze_applications").show()
spark.sql("SELECT COUNT(*) FROM bronze_workflow_events").show() 
```
