```python
from pyspark.sql import functions as F

dim_job = (
    spark.table("bronze_jobs")
    .select(
        F.trim(F.col("job_id")).alias("job_id"),
        F.trim(F.col("title")).alias("title"),
        F.trim(F.col("department")).alias("department"),
        F.col("posted_date").alias("posted_date"),
        F.trim(F.col("status")).alias("status")
    )
    .dropDuplicates(["job_id"])
)

dim_job.write.format("delta").mode("overwrite").saveAsTable("dim_job")
```

```python
spark.sql("SELECT * FROM dim_job LIMIT 10").show()
spark.sql("SELECT COUNT(*) FROM dim_job").show()
```
